```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return "Stack Empty"
    
    def peak(self):
        if not self.is_empty():
            return self.items[-1]
        return "Stack Empty"
    
    def is_empty(self):
        return len(self.items)==0
    
    def size(self):
        return len(self.items)
    

    



    

s = Stack()

s.push(10)
s.push(20)
s.push(30)
print(s.items)
print(s.peak())
print(s.pop())
print(s.peak())
print(s.size())
print()



```

```python
text = "HELLO MOHI"

stack = []

for i in text:
    stack.append(i)
    
reverse = ""

while stack:
    reverse = reverse + stack.pop()
    

print(reverse)
```

```python
num = 13 
stack = []

while num >0:
    stack.append(num%2)
    num = num //2

binary = ""

while stack:
    binary = binary + str(stack.pop())


print(binary)
```
