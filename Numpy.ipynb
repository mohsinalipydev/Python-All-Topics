```python
a = np.array([[1,2],
             [4,5]])
b = np.array([[9,8],
              [4,5]])

result = np.dot(a,b)
print(result)
```

```python
a = np.array([1,0,0])
b = np.array([0,1,0])

cross = np.cross(a,b)
print(cross)

# cross formula
#Cx​ = Ay​Bz​−Az​By​
#Cy=Az​Bx​−Ax​Bz
#Cz​=Ax​By​−Ay​Bx​​
```

Statistical Functions

```python
data = np.array([10,20,30,40])
print(np.var(data))
print(np.std(data))

# concept of variance

# Data: [10, 20, 30, 40]
# Mean = (10+20+30+40)/4 = 25
# Differences from mean = [-15, -5, 5, -15]
# Squared differences = [225, 25, 25, 225]
# Variance = (225+25+25+225)/4 = 125


#standard Deviation concept
#Standard deviation = variance ka square root
```

```python
values = np.array([14,20,35,40,50])

print(np.percentile(values, 50))
print(np.percentile(values, 75))
print(np.percentile(values, 100))

# concept of percentiles
# as you see in code , 50% is 35 
# 75% is 40 
# 100% is 50 
# so you can easily understand what is it mean
```

```python
arr = np.array([1,2,3,4])
print(np.cumsum(arr))

# concept of cumcum
# as you see in code , 
# logic is that 
# 1+2 = 3
# 3+3 = 6
# 6+4 = 10
```
