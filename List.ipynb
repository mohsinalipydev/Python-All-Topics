Concept of append()

append() adds only a single item to the end of a list. If you pass multiple values in the form of a list, it treats the entire list as one item and creates a nested list inside the original list.

If you want to add multiple values individually to a list, you should use the extend() function. extend() adds each element of the given iterable to the list separately, instead of creating a nested list.

```python
list1 = [1,2,3,4]
list1.append(5)
print(list1)
```

```python
list1 = []
list1.append([1,2,3,4])
print(list1)
```

```python
list1 = []
for i in [1,2,3,4,5]:
    list1.append(i)
print(list1)
```

List Comprehension

```python
list1 = [1,2,3,4,5]
square = [i*i for i in list1]
print(square)
```

```python
list1 = [1,2,3,4,5]
for i in list1:
    print([i*i], end=" ")
```

```python
list1 = [1,2,3,4,5,6,7,8]

even = [i for i in list1 if i%2==0]
print(even)
```

```python
list1 = [1,2,3,4,5,6,7,8]
even = [i%2==0 for i in list1]
print(even)
```

Sorting List

```python
list1 = [8,5,0,3,1,4,5]
list1.sort()
print(list1)
```
