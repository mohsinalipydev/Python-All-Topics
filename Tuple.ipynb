We will understand the concept of Tuple in python.

Simple create Tupel

```python
fruits = ("Mangos", "Bananas", "Apple", "Orange")
print(fruits)
print(type(fruits))
```

Access the element from the Tuple

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan")
print(name[0])
print(name[2])
print(name[4])
```

Tuple Slicing

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan")
print(name[1:4])
print(name[0:2])
print(name[1::])
print(name[::1])
```

Find the length of Tuple

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan")

print(len(name))
```

Count the element in Tuple

```python
numbers = (1,2,3,4,5,6,7,7)

print(numbers.count(7))
```

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan","Hassan")
print(name.count("Hassan"))
```

Find the index in Tuple

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan","Hassan")
print(name.index("Ali"))
print(name.index("Aslam"))
```

```python
numbers = (8,7,6,5,4,3)
print(numbers.index(5))
```

Tuple unpacking

```python
person = ("Mohsin", 24, "Multan")
name, age, city = person

print(name)
print(age)
print(city)
```

Loops in Tuple

```python
names = ("Mohsin", "Ali","Hassan","Aslam")
for i in names:
    print(i)
```

```python
numbers = (1,2,3,4,5,5,6)
for i in numbers:
    print(i)
```

```python
names = ("Mohsin","Ali","Aslam","Hassan")
for i in names:
    for j in i:
        print(j,end=" ")
```

Nested Tuple

```python
names = (("Mohsin", "Ali","Hassan"),
         ("Shami","Aslam","Arslan"),
         ("Mohi","Bablu","Ahmed"))

print(names[0])
print(names[2])
print(names[1])
```

```python
names = (("Mohsin", "Ali","Hassan"),
         ("Shami","Aslam","Arslan"),
         ("Mohi","Bablu","Ahmed"))
for i in names:
    for j in i:
        print(j,end=" ")

print(names[0])
print(names[2])
print(names[1])
```

Convert tuple into List : for insert or append new data

```python
name = ("Mohsin","Ali","Hassan","Aslam")

list1 = list(name)
print(list1)
print(type(list1))
```

```python
name = ("Mohsin","Ali","Hassan","Aslam")

list1 = list(name)

list1.append("Gujjar")

name = tuple(list1)

print(name)
```

Join Two Tuples

```python
tpl1 = (1,2,3,4)
tpl2 = (5,6,7,8)

add = tpl1 + tpl2
print(add)
```

this is how we can repeat values in Tuple

```python
data = ("Mohsin",)
print(data *5)
```

```python
data = (1,)
print(data*5)
```

find maximum and minimum value in Tuple

```python
data = (5,3,8,9,4,5)
print(max(data))
print(min(data))
print(sum(data))
```

Check if the value in the Tuple or not

```python
data = ("Mohsin", "Ali","Hassan","Aslam")
if "Mohsin" in data:
    print("Found")
else:
    print("Not Found")
```

```python
data = (1,2,3,4,5,7)
if 8 in data:
    print("Found")
else:
    print("Not Found")
```

make Tuple from User input

```python
data = input("Enter values separated by space")
my_tuple = tuple(data.split())
print(my_tuple)
```
