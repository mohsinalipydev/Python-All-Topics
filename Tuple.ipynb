We will understand the concept of Tuple in python.

Simple create Tupel

```python
fruits = ("Mangos", "Bananas", "Apple", "Orange")
print(fruits)
print(type(fruits))
```

Access the element from the Tuple

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan")
print(name[0])
print(name[2])
print(name[4])
```

Tuple Slicing

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan")
print(name[1:4])
print(name[0:2])
print(name[1::])
print(name[::1])
```

Find the length of Tuple

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan")

print(len(name))
```

Count the element in Tuple

```python
numbers = (1,2,3,4,5,6,7,7)

print(numbers.count(7))
```

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan","Hassan")
print(name.count("Hassan"))
```

Find the index in Tuple

```python
name = ("Mohsin","Ali","Ahmed","Aslam","Hassan","Hassan")
print(name.index("Ali"))
print(name.index("Aslam"))
```

```python
numbers = (8,7,6,5,4,3)
print(numbers.index(5))
```

Tuple unpacking

```python
person = ("Mohsin", 24, "Multan")
name, age, city = person

print(name)
print(age)
print(city)
```

Loops in Tuple

```python
names = ("Mohsin", "Ali","Hassan","Aslam")
for i in names:
    print(i)
```

```python
numbers = (1,2,3,4,5,5,6)
for i in numbers:
    print(i)
```

```python
names = ("Mohsin","Ali","Aslam","Hassan")
for i in names:
    for j in i:
        print(j,end=" ")
```
