Single Line

```python
for i in range(2):
    print("*")
```

Horizontal Line

```python
for i in range(5):
    print("*", end =" ")
```

```python
for i in range(3):
    for j in range(3):
        print(f"i = {i}, j = {j}")
```

```python
for i in range(4):
    for j in range(4):
        print("*", end=" ")
    print()
```

```python
for i in range(1,6):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

```python
for i in range(1,6):
    for j in range(1,6):
        print(i*j, end=" ")
    print()
```

```python
school = ["Class A", "Class B", "Class C"]
student = ["Ali", "Sara", "John"] 
for i in school:
    for j in student:
        print(i, j)
```

```python
for i in range(4):
    for j in range(4):
        print("*", end="  ")
    print()  # Move to the next line after inner loop
    
```
